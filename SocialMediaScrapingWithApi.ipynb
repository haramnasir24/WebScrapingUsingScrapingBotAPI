```python
import requests
import json
from time import sleep

# User credentials
username = 'haramnasir'
apiKey = 'hVnMxPq86QLzn5U1re03KDZhd'

# API Endpoints
apiEndPoint = "http://api.scraping-bot.io/scrape/data-scraper"
apiEndPointResponse = "http://api.scraping-bot.io/scrape/data-scraper-response?"

# Get user input for platform
platform = input("Enter the platform (instagram/twitter/facebook/linkedin): ").strip().lower()

# Determine the scraper and URL/account based on platform
if platform == 'instagram':
    account = input("Enter the Instagram account name: ").strip()
    scraper = 'instagramProfile'
elif platform == 'twitter':
    url = input("Enter the Twitter profile URL: ").strip()
    scraper = 'twitterProfile'
elif platform == 'facebook':
    url = input("Enter the Facebook profile URL: ").strip()
    scraper = 'facebookProfile'
elif platform == 'linkedin':
    url = input("Enter the LinkedIn profile URL: ").strip()
    scraper = 'linkedinProfile'
else:
    print("Unsupported platform")
    exit()


# Create payload
if platform == 'instagram':
    payload = json.dumps({"account": account, "scraper": scraper})
elif platform == 'twitter' or platform == 'linkedin' or platform == 'facebook' :
    payload = json.dumps({"url": url, "scraper": scraper})

headers = {
    'Content-Type': "application/json"
}

# Send POST request
response = requests.request("POST", apiEndPoint, data=payload, auth=(username, apiKey), headers=headers)
if response.status_code == 200:
    response_data = response.json()
    print(response_data)
    responseId = response_data["responseId"]

    # Check response status in a loop
    pending = True
    while pending:
        sleep(5)
        finalResponse = requests.request("GET", f"{apiEndPointResponse}scraper={scraper}&responseId={responseId}", auth=(username, apiKey))
        result = finalResponse.json()
        if isinstance(result, list):
            pending = False
            print(finalResponse.text)
        elif isinstance(result, dict):
            if result.get("status") == "pending":
                print(result["message"])
                continue
            elif result.get("error") is not None:
                pending = False
                print(json.dumps(result, indent=4))
else:
    print(response.text)
```
